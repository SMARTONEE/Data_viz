```python
fig, ax = plt.subplots(2,2)
fig.set_size_inches(15, 15)

#Graphe 2, courbe du profit total réalisé chaque mois. 

ax[0,0].plot(df.month_number, df.total_profit, linewidth=1.0)
ax[0,0].set_title('Profit par mois', fontsize = 15)


#Graphe 3 : même genre mais avec une courbe pour chaque produit. 

ax[0,1].plot(df.month_number, df.facewash, label = "facewash", linewidth=1.0)
ax[0,1].plot(df.month_number, df.facecream, label = "facecream",linewidth=1.0)
ax[0,1].plot(df.month_number, df.toothpaste, label = "toothpaste", linewidth=1.0)
ax[0,1].plot(df.month_number, df.bathingsoap, label = "bathingsoap", linewidth=1.0)
ax[0,1].plot(df.month_number, df.shampoo, label = "shampoo", linewidth=1.0)
ax[0,1].plot(df.month_number, df.moisturizer, label = "moisturizer", linewidth=1.0)
ax[0,1].legend()

ax[0,1].set_title('Ventes par catégories de produits', fontsize = 15)

#Graphe 4 : Nous souhaitons un graphe qui compare, chaque mois, le nombre d'unités vendus pour chaque produit.

width = 0.15  # the width of the bars

# fig, ax = plt.subplots()
rects1 = ax[1,0].bar(df.month_number - 3*width, df.facewash, width, label='facewash')
rects2 = ax[1,0].bar(df.month_number - 2*width, df.facecream, width, label='facecream')
rects3 = ax[1,0].bar(df.month_number - width/2, df.toothpaste, width, label='toothpaste')
rects4 = ax[1,0].bar(df.month_number + width/2, df.bathingsoap, width, label='bathingsoap')
rects5 = ax[1,0].bar(df.month_number + 2*width, df.shampoo, width, label='shampoo')
rects6 = ax[1,0].bar(df.month_number + 3*width, df.moisturizer, width, label='moisturizer')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax[1,0].set_ylabel('Profits')
ax[1,0].set_title('Profits by cat and monthes')
ax[1,0].legend()
ax[1,0].set_xticks(df.month_number)

ax[1,0].bar_label(rects1, padding=100)
ax[1,0].bar_label(rects2, padding=7)
ax[1,0].bar_label(rects3, padding=7)
ax[1,0].bar_label(rects4, padding=7)
ax[1,0].bar_label(rects5, padding=7)
ax[1,0].bar_label(rects6, padding=7)

fig.tight_layout(rect = (0, 0, 1, 1))


#Nous souhaitons un graphe qui donne la contribution de chaque produit dans le profit total (sur l'année étudiée). Merci d'ajouter le pourcentage pour chaque produit.

a = np.array([df.facecream.sum(), df.facewash.sum(), df.toothpaste.sum(), df.bathingsoap.sum(), df.shampoo.sum(), df.moisturizer.sum()])
products = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']

ax[1,1].pie(a, labels = products, autopct="%1.0f%%", explode = (0.1,0.1,0.1,0.15,0.11,0.1))
ax[1,1].set_title('Répartition des ventes par produits en %')
plt.show()

```
